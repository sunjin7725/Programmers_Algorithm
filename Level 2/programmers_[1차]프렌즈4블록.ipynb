```python
import numpy as np
def solution(m, n, board):
    answer = 0
    new_board = [[''] * len(board[0]) for i in range(len(board))]
    for i in range(len(board)):
        for j in range(len(board[0])):
            new_board[i][j] = board[i][j]
    new_board = np.rot90(np.array(new_board),k=3).tolist()
    
    while True:
        delete_set = set()
        for i in range(len(new_board)-1):
            for j in range(len(new_board[i])-1):
                if j >= len(new_board[i+1]) - 1:
                    continue
                if new_board[i][j] == new_board[i][j+1] == new_board[i+1][j] == new_board[i+1][j+1]:
                    for x, y in [(i + 1, j + 1), (i + 1, j), (i, j + 1), (i, j)]:
                        delete_set.add((x, y))

        if len(delete_set) == 0:
            break

        answer += len(delete_set)
        for i, j in reversed(sorted(delete_set)):
            new_board[i].pop(j)
    return answer
```
